## Clean Lyrics before phonemizing

```python
%matplotlib inline
```

```python
import pandas as pd

lyrics_df = pd.read_csv("data/rap_lyrics 2.csv")
lyrics_df = lyrics_df[["title","artist", "lyrics"]]
lyrics_df.rename({"lyrics": "lyrics_g"}, axis=1, inplace=True)
lyrics_df.head()
lyrics_df.describe()
```

## Remove "Embed" from end of each lyric

```python
#apply regex to remove "Embed" from end of each lyric
import re
pattern = r"\d+Embed$"

lyrics_df["lyrics_g"] = lyrics_df["lyrics_g"].str.replace(pattern, "")
```

## Convert to lowercase

```python
lyrics_df["lyrics_g"] = lyrics_df["lyrics_g"].apply(lambda x: x.lower())
```

```python
lyrics_df["lyrics_g"]
```

## Drop Duplicates

```python
print(len(lyrics_df))
lyrics_df.drop_duplicates(inplace=True)
print(len(lyrics_df))
```

## Drop Languages that aren't English 

```python
#! pip3 install langdetect
```

```python
from langdetect import detect

def detect_lyric(x):
    try:
        return detect(x)
    except:
        return ""
        
```

```python
lyrics_df["language"] = lyrics_df["lyrics_g"].apply(detect_lyric)
```

```python
lyrics_df.groupby("language").count()
```

```python
lyrics_df= lyrics_df[lyrics_df.language == "en"]
```

```python
##  Remove " \" and replace ": seperators with "/"
pattern = r"\\"
lyrics_df["lyrics_g"] = lyrics_df["lyrics_g"].str.replace(r"\\",  "")
lyrics_df["lyrics_g"] = lyrics_df["lyrics_g"].str.replace(":",  "/")
```

```python
lyrics_df.head()
```

## Split by "//" to create verses

```python
verses_df = lyrics_df[["artist","title"]]
verses_df["verses_g"] = lyrics_df["lyrics_g"].apply(lambda x: x.split("//"))
```

```python
verses_df = verses_df.explode("verses_g")
display(verses_df)
```

```python
# drop Anuel AA bc it is in Spanish and not caught by lang detect
verses_df = verses_df[verses_df["artist"] != "Anuel AA"]
```

## Drop empty verses

```python
verses_df = verses_df[verses_df["verses_g"]!=""] #drop all empty verses
```

## Filter out verses with minimal lines (bars) or small length

```python
verses_df["length"] = verses_df["verses_g"].apply(lambda x: len(x))
verses_df["bars"] = verses_df["verses_g"].apply(lambda x: len(x.split("/")))
verses_df.describe()
```

```python
verses_df[verses_df["length"]<30]
```

```python
verses_df[verses_df["bars"]<30]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

fig, axes = plt.subplots(1, 2, figsize=(12, 5))


sns.distplot(verses_df[verses_df["length"] <= 500]["length"], kde=True, bins=20, label="length <= 50", ax=axes[0])
axes[0].set_xlabel("verse length")
axes[0].set_ylabel("Density")
axes[0].set_title("Distribution of length (length <= 500)")


sns.distplot(verses_df[verses_df["length"] >5000]["length"], kde=True, bins=20, label="length > 50", ax=axes[1])
axes[1].set_xlabel("verse length")
axes[1].set_ylabel("Density")
axes[1].set_title("Distribution of length (length > 500)")
```

```python
#visually inspect small verses and abnormally large verses
display(verses_df[verses_df["length"] <= 60])
```

```python
#  display(verses_df.sort_values("length",ascending=False))

indices_to_drop = verses_df.nlargest(8, "length").index #drop 8 largest verses because these were too large for model

verses_df.drop(indices_to_drop, inplace=True)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

fig, axes = plt.subplots(1, 2, figsize=(12, 5))


sns.distplot(verses_df[verses_df["bars"] <= 10]["bars"], kde=True, bins=20, label="Bars <= 50", ax=axes[0])
axes[0].set_xlabel("Bars")
axes[0].set_ylabel("Density")
axes[0].set_title("Distribution of Bars (Bars <= 50)")


sns.distplot(verses_df[verses_df["bars"] >100]["bars"], kde=True, bins=20, label="Bars > 50", ax=axes[1])
axes[1].set_xlabel("Bars")
axes[1].set_ylabel("Density")
axes[1].set_title("Distribution of Bars (Bars > 50)")

```

```python
# visually inspect small bar values
display(verses_df[verses_df["bars"] <= 3])
```

```python
# remove verses that are shorter than 55 characters or have three or less bars
verses_df= verses_df[(verses_df["bars"]>3) | (verses_df["length"] >55)]
```

```python
#print versus_df to csv
verses_df = verses_df.reset_index(drop=True)
verses_df.to_csv("data/verses2_cleaned.csv", index=False)
```

# -------------Phonemize in VS Code -------------

```python
# import phoemized dataset and convert to final text file
verses_gp = pd.read_csv("data/verses2_cleaned_phonemized_07-19-2023.csv")
#verses_gp.head

    
```

```python
import random
verses_g = verses_gp["verses_g"].to_list()
verses_p = verses_gp["verses_p"].to_list()

verses_formated = []

for i,verse_g in enumerate(verses_g):
    verses_formated.append(f"<{verse_g} =g2p= {verses_p[i]}>") #append grapheme to phoneme (g2p) task
    verses_formated.append(f"<{verses_p[i]} =p2g= {verse_g}>")  #append phoneme to grapheme (p2g) task
    
random.shuffle(verses_formated)

train_test_split = round(0.99*len(verses_formated))

verses_train = verses_formated[:train_test_split]
verses_test = verses_formated[train_test_split:]

with open("data/train_verses.txt", "w") as f:
    for task in verses_train:
        f.write(task+"\n")
        
with open("data/test_verses.txt", "w") as f:
    for task in verses_test:
        f.write(task+"\n")

display(verses_formated)

```
